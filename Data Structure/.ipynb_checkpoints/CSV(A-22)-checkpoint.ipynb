```python
wf=open("newcsv.csv",'a',newline="")
writer=csv.writer(wf)
writer.writerow(("Name","Addr"))
writer.writerow(("Aastha","ktm"))
writer.writerow(("Ram","bhkt"))
writer.writerow(("Krishna","bnp"))
wf.close()
```

```python
rf=open("newcsv.csv","r")
reader=csv.reader(rf)
for row in reader:
    print(row)
rf.close()   
```

```python
wf=open("new_csv.csv",'a',newline="")
writer=csv.writer(wf)
writer.writerow({"Name":"Addr",
  "Aastha":"ktm",
  "Ram":"bhkt",
  "Krishna":"bnp"})
wf.close()
```

```python
adict=[{"name":"aastha","gender":"female"},
        {"name":"1","gender":"male"}]
```

```python
with open("dict_file.csv","a",newline="") as wf:
    writer=csv.DictWriter(wf,fieldnames=adict[0].keys())
    writer.writeheader()
    writer.writerows(adict)
```
